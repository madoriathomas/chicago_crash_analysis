```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('data/Traffic_Crashes_-_Crashes.csv')
df.head()
```

```python
df.info()
```

```python
to_drop = ['CRASH_RECORD_ID', 'CRASH_DATE_EST_I', 'LANE_CNT', 'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I', 'HIT_AND_RUN_I', 'PHOTOS_TAKEN_I',\
    'STATEMENTS_TAKEN_I', 'DOORING_I', 'WORK_ZONE_I', 'WORK_ZONE_TYPE', 'WORKERS_PRESENT_I', 'LOCATION']

new_df = df.drop(to_drop, axis=1)
new_df.head()
```

```python
new_df['MOST_SEVERE_INJURY'].value_counts()
```

```python
new_df['PRIM_CONTRIBUTORY_CAUSE'].value_counts()
```

```python
new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace(['DISTRACTION - FROM INSIDE VEHICLE',
     'DISTRACTION - FROM OUTSIDE VEHICLE', 'DISTRACTION - OTHER ELECTRONIC DEVICE (NAVIGATION DEVICE, DVD PLAYER, ETC.)', 
     'CELL PHONE USE OTHER THAN TEXTING', 'TEXTING'], 'DISTRACTION/CELL PHONE USE')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace(['DISREGARDING STOP SIGN',
      'DISREGARDING OTHER TRAFFIC SIGNS', 'DISREGARDING ROAD MARKINGS', 'DISREGARDING YIELD SIGN', 
      'BICYCLE ADVANCING LEGALLY ON RED LIGHT', 'MOTORCYCLE ADVANCING LEGALLY ON RED LIGHT'], 'DISREGARDING TRAFFIC SIGNALS')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace(['DRIVING ON WRONG SIDE/WRONG WAY',
     'OPERATING VEHICLE IN ERRATIC, RECKLESS, CARELESS, NEGLIGENT OR AGGRESSIVE MANNER', 'FOLLOWING TOO CLOSELY',
     'IMPROPER OVERTAKING/PASSING', 'IMPROPER BACKING', 'IMPROPER LANE USAGE', 'IMPROPER TURNING/NO SIGNAL', 'PASSING STOPPED SCHOOL BUS',
     'TURNING RIGHT ON RED', 'RELATED TO BUS STOP'], 'DRIVING SKILLS/KNOWLEDGE/EXPERIENCE')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace('NOT APPLICABLE', 'UNABLE TO DETERMINE')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace(['EXCEEDING AUTHORIZED SPEED LIMIT',
     'EXCEEDING SAFE SPEED FOR CONDITIONS', 'FAILING TO REDUCE SPEED TO AVOID CRASH'], 'SPEEDING')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace('ANIMAL', 'EVASIVE ACTION DUE TO ANIMAL, OBJECT, NONMOTORIST')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace(['UNDER THE INFLUENCE OF ALCOHOL/DRUGS (USE WHEN ARREST IS EFFECTED)',
      'HAD BEEN DRINKING (USE WHEN ARREST IS NOT MADE)'], 'UNDER THE INFLUENCE')

new_df['PRIM_CONTRIBUTORY_CAUSE'] = new_df['PRIM_CONTRIBUTORY_CAUSE'].replace(['ROAD ENGINEERING/SURFACE/MARKING DEFECTS', 
      'OBSTRUCTED CROSSWALKS'], 'ROAD CONSTRUCTION/MAINTENANCE')
```
